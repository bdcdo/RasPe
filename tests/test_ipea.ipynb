```python
import requests

url = "https://www.ipea.gov.br/portal/coluna-5/central-de-conteudo/busca-publicacoes"

querystring = {"palavra_chave":"economia verde","tipo":"","assunto":"","autor":"","timeperiods":"all","data-inicial":"","data-final":"^","pagina":"2"}

payload = ""
headers = {
    "cookie": "55b3912ba96bd3dcbe4d325c568ab7e0=528a8fsvqe2a13mtmlvbqun72b",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
    "Accept-Encoding": "gzip, deflate, br, zstd",
    "Accept-Language": "pt-BR,en-US;q=0.7,en;q=0.3",
    "Connection": "keep-alive",
    "DNT": "1",
    "Priority": "u=0, i",
    "Sec-Fetch-Dest": "document",
    "Sec-Fetch-Mode": "navigate",
    "Sec-Fetch-Site": "same-origin",
    "Sec-Fetch-User": "?1",
    "Sec-GPC": "1",
    "Upgrade-Insecure-Requests": "1",
    "User-Agent": "Mozilla/5.0 (X11; Linux x86_64; rv:139.0) Gecko/20100101 Firefox/139.0"
}

response = requests.request("GET", url, data=payload, headers=headers, params=querystring)

print(response.text)
```

```python
from bs4 import BeautifulSoup as bs

req_bs = bs(response.content)
```

```python
lista_infos = []
card_body = req_bs.find('div', class_='lista-publicacoes')
            
if not card_body:
    print('erro')
    
container = card_body.find('div')
if not container:
    print('erro')
    
itens = container.find_all('div', class_="row")

for i in range(len(itens)):
    item = itens[i]
    item2 = item.find('div', class_='publi-conteudo')
    
    link = 'https://www.ipea.gov.br' + item2.find('h3').find('a')['href']
    titulo = item2.find('h3').find('a').text.strip()
    autores = item2.find('div', class_='autores').text.strip()
    data = item2.find('p')
    assuntos = item2.find('div', class_='assuntos').text.strip()

    lista_infos.append([titulo, link, autores, data, assuntos])
```
