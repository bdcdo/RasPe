```python
from legiscraper import scraper

scraper = scraper('PRESIDENCIA')

# Com rastreamento (comportamento padrão)
df = scraper.scrape(pesquisa=["doença rara", "condição rara"])
print(df.head())  # Deve mostrar a coluna termo_busca

# Sem rastreamento
df = scraper.scrape(pesquisa=["doença rara", "condição rara"], rastrear_termo_busca=False)
print(df.head())  # Não deve ter a coluna termo_busca
```
