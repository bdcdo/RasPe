```python
import legiscraper
scraper = legiscraper.scraper("PRESIDENCIA")
```

```python
busca = '(((doença OU doenças OU síndrome OU síndrome OU condição OU condições OU patologia OU patologias) E (rara OU raras OU ultrarrara OU ultrarraras)) OU ((medicamento OU medicamentos OU terapia OU terapias) E (órfão OU órfãos)))'
busca = legiscraper.expand(busca)
```

```python
dados = scraper.scrape(pesquisa=busca, debug=True)
print(dados)
```
